```python
num_1=int(input("Enter the number 1:"))
num_2=int(input("Enter the number 2:"))
if num_1 > num_2:
    print("Number 1 is larger")

elif num_2 > num_1:
    print("Number 2 is larger")

else:
    print("Both numbers are equal")
```
